```python
class StudentGradeAnalysis: #applicable for only one subject data with student name,subject score, Grade (data format)
    def __init__(self):
        self.__file_loaded=False #creating a flag to make sure open_file method is executed first
        self.math_scores=[]
        self.student_names=[]
        self.grades=[]
        self.student_dict={}
        self.student_grade_dict={}

    #to open the student data file
    def open_file(self,filename): 
        """ Takes a string as an argument. Input filename without the extension(txt,pdf etc)"""
        try:
            with open((filename + ".txt"),"r") as file_handler:
                #To seperate the names and scores from the txt file
                
                for student in file_handler.readlines()[1:]:
                    student_data=student.split(",") #spliting into a new list of student names, score, grade
                    self.student_names.append(student_data[0])
                    self.math_scores.append(int(student_data[1]))
                    self.grades.append(student_data[2].strip())
                self.student_dict=dict(zip(self.student_names,self.math_scores))   #student and scores dictionary
                self.student_grade_dict=dict(zip(self.student_names,self.grades))   #student and grade dict
                self.__file_loaded= True
                return f" File loaded Sucessfully"
                
                
        except Exception as e:#if files not found in the folder
            return("something went wrong, Check your file for complete data",e)
            

    #Method for the mean score
    def mean(self):
        """Doesn't take any argument"""
        if not self.__file_loaded: # open_file method has to be executed first before any other method
            return "Load a student grade file using 'open_file()' before using this method"
        return f" Mean score = {round(sum(self.math_scores)/len(self.math_scores),2)}"

    #Minimum score
    def lowest(self):
        """Doesn't take any argument"""
        if not self.__file_loaded:
            return "Load a student grade file using 'open_file()' before using this method"
        elif len(self.math_scores) == len(self.student_names):#to check there's no missing score 
            mini_score= min(self.math_scores)
            lowest_student=[]
            for i in range(len(self.math_scores)): #for two or more students with the same lowestest score
                if self.math_scores[i] == mini_score:
                    lowest_student.append(self.student_names[i])
            return f"{", ".join(lowest_student)} has the lowest score of {mini_score}"
            
        return "Missing Content In The Data file"

    #Maximum score
    def highest(self):
        """Doesn't take any argument"""
        if not self.__file_loaded:
            return "Load a student grade file using 'open_file()' before using this method"
        elif len(self.math_scores) == len(self.student_names):
            max_score= max(self.math_scores)
            highest_student=[]
            for i in range(len(self.math_scores)):  #for two or more students with the same highest score
                if self.math_scores[i] == max_score:
                    highest_student.append(self.student_names[i])
            return f"{", ".join(highest_student)} has the Highest score of {max_score}"
            
        return "Missing Content In The Data file"

    #Method to know how many As,Bs...
    def grade_count(self,grade):
        """Takes a string(letter e.g 'A', 'B' as argument"""
        if not self.__file_loaded:
            return "Load a student grade file using 'open_file()' before using this method"
        return f"{self.grades.count(grade.upper())} Student(s) got '{grade.upper()}'"


    #to get the median score
    def median_score(self):
        """Doesn't take any argument"""
        if not self.__file_loaded:
            return "Load a student grade file using 'open_file()' before using this method"
        sorted_scores= self.math_scores.copy()
        sorted_scores.sort()
        
        #for an even list length
        if len(self.math_scores)%2==0 and len(self.math_scores) == len(self.student_names):
            first=(len(sorted_scores)//2)-1
            second=(len(sorted_scores)//2)
            median=(sorted_scores[first]+sorted_scores[second])/2
            return f" Median Score = {median}"
        
        #for an odd list length
        elif len(self.math_scores)%2 !=0 and len(self.math_scores) == len(self.student_names):
                median_point=len(self.math_scores)//2
                median=sorted_scores[median_point]
                return f" Median Score = {median}"
        return "Missing Content In The Data file"
            
                
    #to check for the most occuring score
    def mode_score(self):
        """Doesn't take any argument"""
        if not self.__file_loaded:
            return "Load a student grade file using 'open_file()' before using this method"
        elif len(self.math_scores) == len(self.student_names):
            mode=max(self.math_scores,key=self.math_scores.count)
            mode_list=[mode]
            if self.math_scores.count(mode)==1:
                return "None"
            else:
                for number in self.math_scores:
                    if self.math_scores.count(number)==self.math_scores.count(mode)and number not in mode_list:
                        mode_list.append(number)
                if len(mode_list)>1:
                    return f" The most common scores = {mode_list}"
                return f" The most common score = {mode}"
        return "Missing Content In The Data file"
        
    
    #to access the grade and score of a student using the student name
    def name_checker(self,student_name):
        """Takes in a full name (separated with a space) or first name, or last name as argument"""
        if not self.__file_loaded:
            return "Load a student grade file using 'open_file()' before using this method"
        striped_name= student_name.strip().lower() # to remove spaces and convert to lower casing
        first_name=[]
        last_name=[]
        striped_student_names=[]
        #To separate full names into first and last name
        for item in self.student_names:
            striped_student_names.append(item.lower())
            first_name.append(item.split()[0].lower())
            last_name.append(item.split()[1].lower())
        
        #if full name is inputed
        if striped_name in striped_student_names and len(self.math_scores) == len(self.student_names):
            full_name_index= striped_student_names.index(striped_name)
            return f"{self.student_names[full_name_index]}, score = {self.math_scores[full_name_index]} grade '{self.grades[full_name_index]}' "
        
        elif len(self.math_scores) == len(self.student_names): 
                
                #if first name is inputed 
                if striped_name in first_name and first_name.count(student_name.strip().lower())==1:
                    first_name_index=first_name.index(striped_name)
                    return f"{self.student_names[first_name_index]}, score = {self.math_scores[first_name_index]} grade '{self.grades[first_name_index]}' "
                
                #if last name is inputed
                elif striped_name in last_name and last_name.count(student_name.strip().lower())==1:
                    last_name_index=last_name.index(striped_name)
                    return f"{self.student_names[last_name_index]}, score = {self.math_scores[last_name_index]} grade '{self.grades[last_name_index]}' "
                    
                #when first name occur more than once in the data
                elif striped_name in first_name and first_name.count(striped_name)>1:
                    all_names=[]
                    for i in range(len(first_name)):
                        if first_name[i] == striped_name:
                            all_names.append(f"{self.student_names[i]}, Score: {self.math_scores[i]}, Grade: '{self.grades[i]}'")
                    return  "  |  ".join(all_names)
                
        
                #when last name occur more than once in the data            
                elif striped_name in last_name and last_name.count(striped_name)>1:
                    all_name=[]
                    for i in range(len(last_name)):
                        if last_name[i] == striped_name:
                            all_name.append(f"{self.student_names[i]}, score = {self.math_scores[i]}, grade '{self.grades[i]}' ")
                    return  "  |  ".join(all_name)
                else:
                    return "Name doesn't exist on the list. Use .student_names to Access Student List libary for Student names "

        return "Missing Content In The Data."

    #to rank the student in terms of their scores
    def score_ranking(self):
        """Doesn't take any argument"""
        if not self.__file_loaded:
            return "Load a student grade file using 'open_file()' before using this method"
        
        #function for the suffix st, nd, rd...
        def suffix(n):
            if 11 <= n % 100 <= 13:
                return f"{n}th"
            last_digit = n % 10
            if last_digit == 1:
                return f"{n}st"
            elif last_digit == 2:
                return f"{n}nd"
            elif last_digit == 3:
                return f"{n}rd"
            else:
                return f"{n}th"
        
        # to sort the dictionary using each value in the dictionary in decending order
        sorted_student_dict=dict(sorted(self.student_dict.items(),key=lambda item: item[1], reverse=True))
        i=0
        prev_value=0
        ranking=[]

        #to assign ranking
        for key,value in sorted_student_dict.items():
            if prev_value== value: #if two or more student have the same scores
                ranking.append(f" {key}, {value}    - {suffix(i)}")
            else:   
                ranking.append(f" {key}, {value}    - {suffix(i+1)}")
            i +=1
            prev_value=value
        return ranking

    
        
                    
    

    
        
```
